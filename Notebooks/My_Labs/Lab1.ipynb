## Подключение библиотек и датасета

```python
import csv
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

relative_path = "diamonds.csv"
absolute_path = os.path.abspath(relative_path)

dataset = pd.read_csv(absolute_path, sep = ',')
dataset.head()
```

---
## 0. Очистка датасета

```python
dataset.drop_duplicates(inplace=True)
dataset.dropna(inplace=True)
dataset.head()
```

---
## 1. Исследование мер центральной тенденции:
   - Найдите среднее значение, медиану и моду для численных переменных.
   - Опишите, что каждая из этих мер говорит о распределении данных.

```python
num_cols = dataset._get_numeric_data().columns

for col in num_cols[1:]:
    print(f"{col}:")
    print(f"Mean: {dataset[col].mean()} \tMedian: {dataset[col].median()} \tMode: {dataset[col].mode().values[0]} \n")
```

---

**Среднее значение**: центральная точка распределения данных (частное суммы значений и их количества)

**Медиана**: середина упорядоченного набора данных

**Мода**: наиболее часто встречающееся значение

## 2. Визуализация данных:
   - Создайте гистограммы для переменных `carat`, `depth`, и `price`.
   - Постройте boxplot для цены в зависимости от качества огранки и цвета алмазов.
   - Используйте scatter plot для изучения взаимосвязи между `carat` и `price`.
  

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.title('Гистограмма для carat')
sns.histplot(data=dataset, x="carat", bins=30)

plt.subplot(2, 2, 2)
plt.title('Гистограмма для depth')
sns.histplot(data=dataset, x="depth", bins=30)

plt.subplot(2, 2, 3)
plt.title('Гистограмма для price')
sns.histplot(data=dataset, x="price", bins=30)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.subplot(1, 2, 1)
sns.boxplot(x='cut', y='price', data=dataset)
plt.title('Boxplot для price по cut')

plt.subplot(1, 2, 2)
sns.boxplot(x='color', y='price', data=dataset)
plt.title('Boxplot для price по color')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.title('Scatter plot для carat и price')
plt.scatter(dataset['carat'], dataset['price'])
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()



# plt.figure(figsize=(12, 8))
# sns.scatterplot(data=dataset, x="carat", y="price")
# plt.show()


# dataset.hist(column='carat')
# dataset.hist(column='depth')
# dataset.hist(column='price')

# dataset.boxplot(column=['price'], by=['cut', 'color'])

# sp = dataset.plot.scatter(x='carat', y='price')
```

## 3. Исследование мер вариативности:
   - Рассчитайте стандартное отклонение и диапазон для переменных `carat`, `depth`, и `price`.
   - Объясните значение этих мер в контексте данных.

```python
for col in ['carat','depth','price']:
    print(f"{col}:")
    print(f"Standard Deviation: {dataset[col].std()}  \t  Range: {dataset[col].max() - dataset[col].min()} \n")
```

---

**Стандартное отклонение**: позволяет увидеть насколько сильно значения отклоняются от среднего

**Диапазон**: разница между максимальным и минимальным значением

## 4. Корреляционная матрица:
   - Создайте корреляционную матрицу для всех численных переменных.
   - Используйте тепловую карту для визуализации корреляционной матрицы.
   - Обсудите, какие переменные наиболее сильно коррелируют и предположите почему.

```python
corr_matrix = dataset[num_cols[1:]].corr()
plt.figure(figsize=(10, 8))
plt.title('Корреляционная матрица')
sns.heatmap(corr_matrix, annot=True, cmap="YlGnBu")
plt.show()
```

```python
mask_matrix = np.triu(dataset[num_cols[1:]].corr())
plt.figure(figsize=(10, 8))
plt.title('Корреляционная матрица')
sns.heatmap(corr_matrix, annot=True, cmap="YlGnBu", mask=mask_matrix)
plt.show()
```
